```python
# Example of using enumerate() with a list
my_list = ['apple', 'banana', 'cherry', 'date']

for index, value in enumerate(my_list):
    print(f"The item at index {index} is {value}")

# Example of using enumerate() with a starting index
for index, value in enumerate(my_list, start=1):
    print(f"The item at position {index} is {value}")

# Using enumerate to create a dictionary
my_dict = {index: value for index, value in enumerate(my_list)}
print(my_dict)
```

```python
import re

# Example of using re.match()
pattern = r"hello"
string = "hello world"
match = re.match(pattern, string)
if match:
    print("Match found")
else:
    print("No match found")

# Example of using re.search()
pattern = r"world"
string = "hello world"
search = re.search(pattern, string)
if search:
    print("Pattern found")
else:
    print("Pattern not found")

# Example of using re.findall()
pattern = r"apple"
string = "apple banana apple cherry apple"
matches = re.findall(pattern, string)
print("Matches found:", matches)

# Example of using re.sub()
pattern = r"apple"
string = "apple banana apple cherry apple"
replacement = "orange"
new_string = re.sub(pattern, replacement, string)
print("New string:", new_string)
```
